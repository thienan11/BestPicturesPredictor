# Machine Learning

This notebook contains code for the machine learning model that we chose as well as the our testing anaylsis of the model

```python
import pandas as pd
```

```python
df_bp = pd.read_csv("oscars_best_pictures.csv")

df_bp = df_bp.drop(columns=['Unnamed: 0'])

df_bp
```

##Removing NaN Rows

```python
df_bp.dropna(subset=["imdb_rating", "rt_critic", "rt_audience"], inplace=True)
df_bp[df_bp["year_ceremony"] == 1949]
```

```python
df_bp.reset_index(drop=True, inplace=True)
df_bp
```

##Setup

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score

X_train = df_bp[(df_bp["year_ceremony"] >= 1949) & (df_bp["year_ceremony"] < 2023)]
y_train = df_bp[(df_bp["year_ceremony"] >= 1949) & (df_bp["year_ceremony"] < 2023)]

X_test = df_bp[(df_bp["year_ceremony"] >= 2023)]
y_test = df_bp[(df_bp["year_ceremony"] >= 2023)]

indices = X_test.index

X_train = X_train.drop(['winner', 'runtime', 'year_film', 'year_ceremony', 'ceremony', 'category', 'name', 'film',
                'director', 'stars', 'gross', 'metascore'], axis=1)
y_train = y_train['winner'].astype(int)

X_test = X_test.drop(['winner', 'runtime', 'year_film', 'year_ceremony', 'ceremony', 'category', 'name', 'film',
                'director', 'stars', 'gross', 'metascore'], axis=1)
y_test = y_test['winner'].astype(int)

ct = make_column_transformer(
    (StandardScaler(), ["total_noms", "imdb_rating", "rt_critic", "rt_audience"])
)
```

## Models

###K-Nearest Neighbor Classifier

####Fitting KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

np.random.seed(42)

knn_model = make_pipeline(
    ct,
    KNeighborsClassifier(n_neighbors=5)
)

knn_model.fit(X_train, y_train)
```

####Calculating Best Features for KNN

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(knn_model,
                           param_grid={
                               "kneighborsclassifier__n_neighbors": range(1, 20),
                               "kneighborsclassifier__metric": ["euclidean", "manhattan"]
                           },
                           scoring="neg_mean_squared_error",
                           cv=10)

grid_search.fit(X_train, y_train)
grid_search.best_estimator_
pd.DataFrame(grid_search.cv_results_)
```

```python
knn_model = make_pipeline(
    ct,
    KNeighborsClassifier(n_neighbors=17)
)

knn_model.fit(X_train, y_train)
```

###Logistic Regression

####Fitting Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = make_pipeline(
    ct,
    LogisticRegression(random_state=42)
)

lr_model.fit(X_train, y_train)
```

####Hyperparameter Tuning

```python
param_grid = {
    'logisticregression__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'logisticregression__solver': ['liblinear', 'saga', 'newton-cg', 'lbfgs'],
    'logisticregression__max_iter': [100, 500, 1000, 5000]# Choose solvers
}

grid_search = GridSearchCV(lr_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
lr_model = make_pipeline(
    ct,
    LogisticRegression(random_state=42, C=1, max_iter=100, solver="liblinear")
)

lr_model.fit(X_train, y_train)
```

###Random Forest Classifier

####Fitting Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = make_pipeline(
    ct,
    RandomForestClassifier(random_state=42)
)

rf_model.fit(X_train, y_train)
```

####Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'randomforestclassifier__n_estimators': [100, 300, 500],
    'randomforestclassifier__max_depth': [5, 10, 20],
    'randomforestclassifier__min_samples_split': [2, 10, 20]
}

grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
rf_model = make_pipeline(
    ct,
    RandomForestClassifier(random_state=42, n_estimators=500, min_samples_split=20, max_depth=5)
)

rf_model.fit(X_train, y_train)
```

### Stacking Classifier

####Fitting Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier

ensemble_model = StackingClassifier([
    ("knn", knn_model),
    ("linear", lr_model)
])

ensemble_model.fit(X_train, y_train)
```

##Model Testing and Analysis

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np
```

####K-Nearest Neighbor Analysis

```python
scores = cross_val_score(knn_model,
                         X=X_train,
                         y=y_train,
                         scoring="accuracy",
                         cv=5)
scores.mean()
```

#####Accuracy, Precision, Recall, F1, & Confusion Matrix

```python
y_train_pred = knn_model.predict(X_train)
y_train_prob = knn_model.predict_proba(X_train)[:, 1]

train_accuracy = knn_model.score(X_train, y_train)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

print(f'Training Accuracy: {train_accuracy}')
print(f'Training Precision: {train_precision}')
print(f'Training Recall: {train_recall}')
print(f'Training F1 Score: {train_f1}')

train_conf_matrix = confusion_matrix(y_train, y_train_pred)
print('Training Confusion Matrix:\n', train_conf_matrix)
```
